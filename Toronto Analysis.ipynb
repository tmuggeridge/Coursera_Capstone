```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
dfs = pd.read_html(url)
df = dfs[0]
df2 = df[['Postal Code','Borough','Neighbourhood']]
indexBorough = df2[ df2['Borough'] == 'Not assigned' ].index
df2.drop(indexBorough , inplace=True)
df3 = df2.groupby('Postal Code').agg( ','.join)
df3
```

```python
url2 = 'https://cocl.us/Geospatial_data'
coordinates = pd.read_csv(url2)
coordinates
```

```python
combined = pd.merge(df3, coordinates, on = 'Postal Code')
combined
```

Below I shall remove the categorical variables and then fit the kmeans clustering model

```python
latlong = combined.drop(['Postal Code','Borough','Neighbourhood'], axis=1)
latlong.head()
```

```python
# set number of clusters
kclusters = 5

Toronto_clustering = latlong

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Toronto_clustering)
labels = kmeans.labels_

print(labels)
```

```python
labelsdf = pd.DataFrame(labels)
labelsdf.value_counts()
```

The above shows us that most postal codes are grouped in cluster 3, with the fewest grouped in cluster 2
